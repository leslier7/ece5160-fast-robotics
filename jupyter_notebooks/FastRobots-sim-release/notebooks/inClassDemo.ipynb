```python
# Reset Plotter and Simulator
cmdr.reset_plotter()
cmdr.reset_sim()

cmdr.plot_map()

# Loop for pose
while cmdr.sim_is_running() and cmdr.plotter_is_running():
    pose, gt_pose = cmdr.get_pose()

    cmdr.plot_odom(pose[0], pose[1])
    cmdr.plot_gt(gt_pose[0], gt_pose[1])
```

```python
# Loop for sensor
while cmdr.sim_is_running() and cmdr.plotter_is_running():
    sensor_values = cmdr.get_sensor()
    print(sensor_values)
    await asyncio.sleep(1)
```

## Closed loop control

```python
await asyncio.sleep(1)

while cmdr.sim_is_running():
    sensor_values = cmdr.get_sensor()
    #print(sensor_values)
    if sensor_values[-1] < 0.7:
        cmdr.set_vel(0, 5)
    else:
        cmdr.set_vel(1,0)
    await asyncio.sleep(0.01)
print("Stopping closed loop control")
```
