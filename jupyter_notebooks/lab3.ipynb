## Calibration

```python
from matplotlib import pyplot as plt
import numpy as np

# format: [actual_cm, measured_cm]
data = np.array([
    [7, 6.8],
    [16, 16.3],
    [30, 29.9],
    [40, 41.2],
    [61, 58],
    [100, 97],
    [120, 115]
])

actual = data[:, 0]
measured = data[:, 1]

# Compute error
error = measured - actual
mean_error = np.mean(error)
rmse = np.sqrt(np.mean(error**2))

# Plot measured vs actual
plt.figure()
plt.scatter(actual, measured)
plt.plot(actual, actual)  # Ideal y = x line

plt.xlabel("Actual Distance (cm)")
plt.ylabel("Measured Distance (cm)")
plt.title("Measured vs Actual Distance")

plt.grid(True)
plt.legend(["Measured Data", "Ideal (y=x)"])

plt.show()

print(f"Mean Error: {mean_error:.3f} cm")
print(f"RMSE: {rmse:.3f} cm")
```
